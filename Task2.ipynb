```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
data1 = pd.read_csv("test.csv")
```

```python
data1.head()
```

```python
data1.describe()
```

```python
data1.info()
```

```python
data1.isnull().sum()
```

```python
data1.dropna(subset=["Embarked"], inplace=True)
data1["Cabin"].fillna("Unknown",inplace=True)
data1["Age"].fillna(data1["Age"].mean(),inplace=True)
```

```python
data1.isnull().sum()
```

```python
data1.duplicated().sum()
```

```python
plt.figure(figsize = (10,6))
sns.histplot(data1["Age"],kde=True)
plt.title("Distribution using age")
plt.xlabel("Age")
plt.ylabel("Count")
plt.show()
```

```python
plt.figure(figsize = (10,6))
sns.countplot(data = data1, x = "Sex", hue="Sex")
plt.title("Gender based Survival")
plt.xlabel("Gender")
plt.ylabel("Count")
plt.show()
```

```python
plt.figure(figsize = (10,6))
sns.scatterplot(data = data1, x = "Age", y ="Fare", hue="Sex")
plt.title("Scatter plot of Gender and Fare")
plt.xlabel("Age")
plt.ylabel("Fare")
plt.show()
```
